```python
df.groupby('SIGLA DA UF').agg(lambda x: x.max(u'NÚMERO DE PARTICIPANTES NO ENEM'))
```

```python
df.loc[df.groupby('SIGLA DA UF')[u'NÚMERO DE PARTICIPANTES NO ENEM'].idxmax(), :]\
[['SIGLA DA UF', 'NOME DA ENTIDADE', u'NÚMERO DE PARTICIPANTES NO ENEM']]
```

```python
df.loc[df.groupby('SIGLA DA UF')[u'TAXA DE REPROVAÇÃO'].idxmax(), :]\
[['SIGLA DA UF', 'NOME DA ENTIDADE', u'TAXA DE REPROVAÇÃO']]
```
