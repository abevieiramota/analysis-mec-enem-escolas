```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('../Dataset/full.csv', encoding='utf-8', index_col='id')
```

```python
df['indicador_nivel_socioeconomico'] = df.indicador_nivel_socioeconomico.astype('category', \
categories=[u'Sem informa\xe7\xe3o', u'Muito Baixo', u'Baixo', u'M\xe9dio Baixo', u'M\xe9dio', u'M\xe9dio Alto', 
 u'Alto', u'Muito Alto'], ordered=True)
```

```python
media_columns = ['media_CH', 'media_RED', 'media_MAT', 'media_CN', 'media_LC']
```

```python
_uf = df.groupby('uf')[media_columns].agg(['mean', 'min', 'max'])
```

```python
_i_se = df.groupby('indicador_nivel_socioeconomico')[media_columns].agg(['mean', 'min', 'max'])
```

```python
_porte = df.groupby('porte')[media_columns].agg(['mean', 'min', 'max'])
```

# Mean

```python
_uf.swaplevel(0, 1, axis=1)['mean']\
.plot(kind='bar', subplots=True, figsize=(16, 16), sharex=False, layout=(3, 2), sharey=True)
```

```python
_i_se.swaplevel(0, 1, axis=1)['mean']
```

```python
_i_se.swaplevel(0, 1, axis=1)['mean']\
.plot(kind='bar', rot=80, subplots=True, figsize=(14, 8), sharex=False, layout=(1, 5), sharey=True)
```

```python
_porte.swaplevel(0, 1, axis=1)['mean']\
.plot(kind='bar', rot=80, subplots=True, figsize=(14, 8), sharex=False, layout=(1, 5), sharey=True)
```
