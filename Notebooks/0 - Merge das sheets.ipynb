```python
ds_names = ['CH', 'CN', 'LC', 'MAT', 'RED']
column_id = ['id']
columns_escola = ['nome', 'uf', 'municipio', 'dependencia_administrativa', 'localizacao',\
               'n_alunos_censo', 'porte', 'n_alunos_participantes_enem', 'taxa_participacao',\
               'n_participantes_necessidades_especiais', 'indicador_permanencia_escola',\
               'indicador_nivel_socioeconomico', 'faixa_indicador_formacao_docente',\
               'indicador_formacao_docente', 'taxa_aprovacao', 'taxa_reprovacao',\
               'taxa_abandono']
columns_desempenho = ['media_30_melhores', 'media', 'pct_alunos_nivel_1',\
               'pct_alunos_nivel_2','pct_alunos_nivel_3','pct_alunos_nivel_4',\
               'pct_alunos_nivel_5']

columns_names = column_id + columns_escola + columns_desempenho

dfs = {ds_name: pd.read_csv('..\Dataset\enem_escola_2014_%s.csv' % ds_name, encoding='utf-8', names=columns_names, index_col=0, skiprows=1) for ds_name in ds_names} 
```

```python
# assunção: todas as sheets possuem informações sobre o mesmo conjunto de escolas
# TODO: validar
escolas_df = dfs['CH'][columns_escola]
```

```python
# removendo as colunas de escola dos dfs
for df in dfs.values():
    
    df.drop(columns_escola, axis=1, inplace=True)
```

```python
full = escolas_df

for ds_name, df in dfs.iteritems():
    
    df.columns = ['%s_%s' % (column, ds_name) for column in df.columns]
    
    full = pd.merge(full, df, left_index=True, right_index=True)
```
